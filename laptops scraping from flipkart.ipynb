```python
from bs4 import BeautifulSoup
import requests
url1= 'https://www.flipkart.com/search?q=laptop&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
response1 = requests.get(url1)
```

```python
flipkart=BeautifulSoup(response1.content,"html5lib")
print(flipkart)
```

```python
laptop_names=[]
price=[]
ratings=[]
specification=[]
for i in flipkart.findAll('div',attrs={'class':'_4rR01T'}):
    laptop_names.append(i.text)
    print('laptop_names',laptop_names)
for i in flipkart.findAll('div',attrs={'class':'_25b18c'}):
    price.append(i.text[0:8])
for i in range(len(price)):
    if len(price)>24:
        price.pop(len(price)-1)
print(price)
for i in flipkart.findAll('div',attrs={'class':'_3LWZlK'}):
    ratings.append(i.text)
for i in range(len(ratings)):
    if len(ratings)>24:
        ratings.pop(len(ratings)-1)
print(ratings)
for i in flipkart.findAll('div',attrs={'class':'fMghEO'}):
    specification.append(i.text)
for i in range(len(specification)):
    if len(specification)>24:
        specification.pop(len(specification)-1)
print(specification)
print(len(specification))
```

```python
import pandas as pd
laptops=pd.DataFrame({'laptop names':laptop_names,'Price':price,'Ratings':ratings,'Specifications':specification})
laptops['Price']=laptops['Price'].str.replace("₹","")
laptops
```

```python
laptops.to_csv('Laptops.csv')
```
